```python
# Crear un nuevo Jupyter Notebook llamado graficos.ipynb
!touch graficos.ipynb

# CÃ³digo para graficar los datos en graficos.ipynb
import matplotlib.pyplot as plt

# Leer los datos del archivo datos.txt
with open('datos.txt', 'r') as file:
    lines = file.readlines()

voltages = []
rewards = []
```

```python
# Extraer los valores de Voltage y Reward
for line in lines:
    if line.startswith('Voltage:'):
        parts = line.split(',')
        voltage = float(parts[0].split(':')[1].strip())
        reward = float(parts[4].split(':')[1].strip())
        voltages.append(voltage)
        rewards.append(reward)

# Crear el grÃ¡fico
plt.figure(figsize=(10, 6))
plt.plot(voltages, rewards, marker='o', linestyle='-', color='b')
plt.title('Voltage vs Reward')
plt.xlabel('Voltage')
plt.ylabel('Reward')
plt.grid(True)
plt.savefig('graficos.png')
plt.show()
```
